```python
#Add or remove libraries as needed
import cv2
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

### Task#1: Load/read cat.jpeg as a grayscale image [1 mark]

```python
 img = cv2.imread('cat.jpeg', 0)
plt.imshow(img, cmap='gray')
plt.axis('off')
```

### Task#2: Apply thresholding (thresh=100) to the image and display the results [1.5 mark]
#### Note: use cmap = 'gray' to properly display the thresholded image

```python
#Task2

plt.subplot(1,2,1)
plt.imshow(img, cmap='gray')
plt.title('Original')
plt.axis('off')



_, th = cv2.threshold(img, 100, 255, cv2.THRESH_BINARY)

 

plt.subplot(1,2,2)
plt.imshow(th, cmap='gray')
plt.title('Threshold')
plt.axis('off')
```


### Task#3: Load/read the camera man, blur it using a 9x9 gaussian filter (sigma=5), and display the results.   [2.5 marks]

```python
#Task3

camera = cv2.imread('cameraman.tif', 0)

blur = cv2.GaussianBlur(camera, (9,9), 5)


plt.subplot(1,2,1)
plt.imshow(camera, cmap='gray')
plt.title('original')
plt.axis('off')


plt.subplot(1,2,2)
plt.imshow(blur, cmap='gray')
plt.title('blur')
plt.axis('off')
```
