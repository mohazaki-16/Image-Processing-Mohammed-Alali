```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Task1
#Increase the size of the image

img = cv2.imread('images/cameraman.tif')
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')
plt.figure(figsize=(2, 2))
plt.show()


scaled_img = cv2.resize(img, None, fx=2, fy=2)

plt.figure(figsize=(6,6))
plt.imshow(scaled_img)
plt.title('Scaled Image (2x)')
plt.axis('off')
plt.show()
```

```python
#Rotate the image by 120 degrees

img = cv2.imread('images/cameraman.tif')
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')
plt.figure(figsize=(2, 2))
plt.show()


h,w = img.shape[:2] # so we take two chanels only
center = (w//2, h//2)

M = cv2.getRotationMatrix2D(center, 120, 1) # to create the rotation matrix

rotated = cv2.warpAffine(img, M, (w, h)) # to do the rotation

plt.imshow(rotated)
plt.title("Rotated 120 Degrees")
plt.axis("off")
plt.show()
```

```python
#Perform Sheer operations over the image

img = cv2.imread('images/cameraman.tif')
plt.imshow(img)
plt.title('Original Image')
plt.axis('off')
plt.figure(figsize=(2, 2))
plt.show()


sh = 0.5

#shear matrix (X-direction)
M = np.float32([[1, sh, 0],
                [0, 1, 0]])

sheared = cv2.warpAffine(img, M, (int(w + sh*h), h)) # to do the shear matrix

plt.imshow(sheared)
plt.title("X-Shear")
plt.axis("off")
plt.show()

```
