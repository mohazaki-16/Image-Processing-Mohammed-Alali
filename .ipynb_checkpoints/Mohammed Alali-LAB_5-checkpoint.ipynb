```python
import cv2
import matplotlib
import matplotlib.pyplot as plt
import numpy as np
from skimage import data, img_as_float
from skimage import exposure
from PIL import Image
```

```python
image = data.chelsea()

#Doing 7x7 box filter
smoothed = cv2.blur(image, (7, 7))

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")
plt.subplot(1,2,2)
plt.imshow(smoothed)
plt.title("7x7 Box Filter")
plt.axis("off")

plt.show()
```

```python
image = Image.open ('images/cameraman.tif')
image = np.array(image) 

gaussian_5 = cv2.GaussianBlur(image, (5,5), 0)
gaussian_21 = cv2.GaussianBlur(image, (21,21), 0)

plt.figure(figsize=(15,5))

plt.subplot(1,3,1)
plt.imshow(image)
plt.title("Original Image")
plt.axis("off")

plt.subplot(1,3,2)
plt.imshow(gaussian_5)
plt.title("5x5 Gaussian")
plt.axis("off")

plt.subplot(1,3,3)
plt.imshow(gaussian_21)
plt.title("21x21 Gaussian")
plt.axis("off")

plt.show()
```
